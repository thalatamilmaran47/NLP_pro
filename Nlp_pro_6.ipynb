```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng') # Added as suggested by the error message

sentence = "My name is Tamilmaran. I am a student and I am currently pursuing a degree in Computer Science. I am very interested in learning new technologies and improving my skills in programming and problem-solving. I am a hardworking, self-motivated, and positive person. I enjoy learning new things and working both independently and as part of a team. My goal is to build a successful career and contribute positively to society."

tokens = word_tokenize(sentence)

pos_tags = pos_tag(tokens)

print("Word\t\tPOS Tag\t\tGrammatical Role")
print("-"*55)

pos_meaning = {
    'NN': 'Noun',
    'NNS': 'Plural Noun',
    'NNP': 'Proper Noun',
    'VB': 'Verb',
    'VBD': 'Past Tense Verb',
    'VBG': 'Verb (Gerund)',
    'VBN': 'Past Participle',
    'VBP': 'Present Verb',
    'VBZ': 'Verb (3rd Person)',
    'JJ': 'Adjective',
    'RB': 'Adverb',
    'PRP': 'Pronoun',
    'IN': 'Preposition',
    'DT': 'Determiner'
}

for word, tag in pos_tags:
    role = pos_meaning.get(tag, "Other")
    print(f"{word}\t\t{tag}\t\t{role}")
```
